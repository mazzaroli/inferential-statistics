## Estadística inferencial vs. descriptiva

Cual es la diferencia entre estadistica descriptiva y estadistica inferencia?

Recordando lo que vimos en el articulo de [estadistica descriptiva]().

La estadistica descriptiva nos ayuda a describir, entender y explicar como se estan comportando los datos en el **presente**, lo cual nos basamos en la informacion clave que disponemos en nuestra tabla de datos, donde limpiamos los datos y sacamos indices estadisticos principales

Por otro lado en la estadistica inferencial buscaremos sacar inferencias, inducciones, trataremos de predicir. entenderemos que podemos obtener de estos modelos y validar teorias

Estadistica descriptiva

Nos ayuda a determinar como se comportan los datos:

- La **tendencia** dentro de una poblacion o variable

- La **variabilidad** cual es la dispersion de la problacion o variable

- La **distribucion** de una variable donde veremos si es unimodal, bimodal, multimodal, etc

- Ver los **estadisticos principales** com ola media, mediana y la moda

Estadistica inferencial

Veremos la abstraccion de los datos, donde partiremos de una muestra de una poblacion sobre la cual nos haremos las preguntas y trabajaremos:

- Tecnicas de **muestreo**
- **Intervalos de confianza** para sacar conclusiones y deducciones
- **Validacion de hipotesis** que seran teorias y las demostraremos
- Tecnnicas para **evitar sesgos**

Cuando hablamos de inferencia estadistica nos referimos acerca de sacar **conclusiones que se obtienen a partir de parametros que tenemos de una poblacion de datos** de diferentes puntos de informacion.

Ademas estudiamos que grado de **fiabilidad obtenemos de nuestros resultados, para concluir si nuestro modelo es fiable**, o deberemos recalcular el modelo.

- Entender la **distribucion** de la informacion, ver como se comportan nuestros datos

## Estadisticos principales

Cuales son los componentes principales de la estadistica?

A estos componentes los llamamos estadisticos

### Experimento

El experimento es un procedimiento que puede suceder una vez, varias veces o infinitamente, donde podemos esperar obtener el mismo resultado o distintos resultados cada vez que ejecutemos el experimento. 

Experimento aleatorio: es el caso en que el experimento tiene mas de un resultado posible. Ej: Lanzar un dado, donde el resultado puede ser una de las 6 caras por cada lanzamiento.

Experimento determinista: es el caso en que solo obtenemos un resultado del experimento. Ej: si caminamos a una velocidad constante, ir del punto A al punte B siempre nos tomaria el mismo tiempo.

### Poblacion y muestra


**Poblacion:** es nuestro conjunto total de individuos, objetos o eventos que tienen la misma caracteristica sobre el que estamos interesados obtener conclusiones.\
Ej: los habitantes de Cordoba Argentina 

**Muestra:** es un subconjunto de la poblacion, la cual se slecciona con el proposito de obtener informacion\
Ej: los habitantes *jovenes*, los habitantes *adultos*, etc.

<img src='https://diferenciando.com/wp-content/uploads/2019/09/Opt-0004-Tamao-de-muestra.jpg' width=400>

Que tener en cuenta al crear una muestra?

- La muestra **tiene que tener un numero lo suficiente de registros** para ser **estadisticamente significante** para poder sacar una conclusion. Ej: de una poblacion de 10.000.000 de habitantes sacamos una muestra de 100 habitantes, probablemente nuestra conclusion sea erronea por falta de datos. [Calculadora de muestras](https://www.questionpro.com/es/calculadora-de-muestra.html)

- La muestra tiene que ser una representacion **no sesgada** de la informacion. Ej: Si trabajamos

### Evento o suceso

Los **eventos o sucesos** son los **resultados que obtenemos tras haber realizado el experimento**, por ejemplo de haber lanzado una moneda obtendriamos cara o cruz, en caso de un dado 

### Variable o feature

Las variables o features son las caracteristicas o atributos que tenemos de la poblacion o la muestra, por ejemplo las caracteristicas que varian en una persona serian el color del pelo, peso, estatura, etc.

Tipos:

- Cualitativos: son aquellos atributos categoricos (no medibles). Ej: color de pelo (negro) o el color de ojos (verde).

- Cuantitativos: son aquellos atributos numericos (medibles), se dividen en discretos y contunios
    - Discretos: son aquellos que son un numero entero, por ejemplo cuantas veces compramos un producto, ya que no puede tomar el valor de 1.5 veces e iria de 1, 2, 3 o mas veces
    
    - Continuos: son aquellos que son un numero con decimales, por ejemplo la altura (180.7 cm), peso (122.4 kg), etc

### Probablidad

Es la capacidad de medir **que tan posible** es que ocurra un evento o suceso determinado, por ejemplo al lanzar una moneda tendremos una cierta probabilidad para que caiga en cara o en cruz.

Y la **estadistica** es la ciencia por la cual analizaremos los eventos para saber que tan probables son sus sucesos.

## Poblaciones normales

### Distribucion normal

Logramos apreciar en la naturaleza, que una gran cantidad de sus factores se rigen en base a una poblacion normal. 

Una problacion normal que es aquella que tiende a tener una **distribucion normal o gaussiana**, que de manera grafica la podemos ver en el plano de ejes cartesianos con **forma de campana**, donde la gran mayoria de **los datos estan concentrados en el centro del grafico**, a mismo modo vemos que hay una concentracion de datos mucho menor en los extremos izquierdo y derecho del grafico, estos datos serian concentraciones de **valores atipicos** (outliners en ingles) de nuestra distribucion normal.

En este tipo de distribuciones existe un factor en comun donde la **moda, media y mediana** representan al **mismo valor**

- Distribucion normal = distribucion de Gauss
- Moda = media = mediana
- La distribucion de los datos es simetrica
- Tiene forma de campana

### Ejemplos de poblacion normal

#### 1. Calorias ingeridas y peso

Un ejemplo de como vemos una distribucion normal en la vida cotidiana es el efecto de calorias ingeridas y como afecta al peso de las personas.

La mayoria de las personas tienen una dieta de al rededor de 2000 a 2500 calorias diarias, la cual centra a la mayoria de las personas en el centro de la campana, y por lo tanto la poblacion normal de personas tienden a tener un peso similar.

Pero en el caso de que una persona aumente el numero de calorias ingerida, se tendria a desplazarse al lado derecho de la campana, donde tambien aumentaria su peso en comparacion a la poblacion normal.

A misma manera si una persona come por debajo del promedio, se desplazaria a la izquierda del grafico siendo tambien un valor atipico.

#### 2.Presion sanguinea

Otro ejemplo es el caso de la presion sanguinea, donde la mayoria de las personas tienen una presion similar en la distribucion de la misma, sin embargo hay personas que tienen una preison sanguinea mas baja que la poblacion posicionandose a la izquierda de la distribucion, a su mismo modo las personas que tengan una presion sanguinea mas alta se posicionarian mas a la derecha de la distribucion

#### 3. Tamanos de objetos producidos por una maquina

Cuando nos referimos a la automatizacion de procesos industriales, por ejemplo donde una maquina fabrica un componente de un producto, por lo general este producto tendra siempre el mismo tamano, sin embargo los productos atipicos tendran un tamano mas pequeno o mas grande de lo que debia obtener de la produccion, posicionandose o mas a la izquierda o mas a la derecha que los productos de un tamano normal.

## Introduccion al muestreo y teorema central del limite

Veamos mas sobre el tipo de muestro, como crear estas muestras y tambien  sobre el teorema del limite central

### Muestreo

Como ya sabemos las muestras son una extraccion de una poblacion general, recordemos que debmos asegurarnos que cumplan dos requisitos, que sea lo suficientemente grande para poder obtener una conclusion osea **estadisticamente significativo** y la segunda condicion es que no este sesgada, es decir que la muestra posea diferentes atributos y no nos concentremos en un punto especifico de la poblacion.

El muestreo:

- Es una tecnica para la seleccion dfe una muestra. 

- Se obtiene a partir de una poblacion estadistica.

- La seleccion tiene que ser aleatoria y se espera que sus propiedades sean extrapolables a la poblacion.

### Tipos de muestreos principales

#### Aleatorio simple

Metodo de seleccion de ciertas unidades sacadas de una poblacion de manera que cada una de las muestas tiene las misma probabildiad de seer elegida.

Ejemplo: loteria

#### Sistematico

Metodo de seleccion de ciertas unidades al azar y, a continuacion, se eligen el resto siguiendo intervalos regulares por condiciones internas por condiciones exogenas.

Ejemplo: 

- Dar un precio cada cien persoans quie hagan una inscripcion hasta llegar a un total de mil inscritos.

- Dar un precio a la primera persona que mande un mensaje cuando llueva en un lugar en particular.

#### Estratificado

Metodo de seleccion de ciertas unidades por segmentos exclusivos y homogeneos y, a continuacion se elige una muestra aleatoria simple de cada segmento

Ejemplo: hacer un muestreo de la poblacion Espanola donde se extraigan de manera aleatoria a todos los jovenes de origen argentino.

### Teorema del limite central

> **El teorema central del límite** (TCL) es una teoría estadística que establece que, **dada una muestra aleatoria suficientemente grande** de la población, **la distribución** de las medias muestrales **seguirá una distribución normal**. [Fuente](https://economipedia.com/definiciones/teorema-central-del-limite.html)

El teorema del límite central demuestra que la media de diferentes muestras obtenidas de una misma distribución (ya sea una distribución normal, exponencial o gamma) van a a darnos una distribución normal.

Es decir., si de una población, tomamos varias muestras, y de cada una de esas muestras calculamos la media. el conjunto de esas medias va a describir una distribución normal.

Mientras más muestras mas evidente es esa distribución.

Lo importante aquí es entender que no importa el tipo de distribución de esa población. Las medias de esas muestras siempre me darán una distribución normal. Eso demuestra el teorema del límite central .

[StatQuest: ¡El teorema del límite central, claramente explicado!](https://www.youtube.com/watch?v=YAlJCEDH2uY)

**Ejemplo Codigo**

````python
# Dependencias
import numpy as np
import matplotlib.pyplot as plt


def normal_dist(k, mean, std):
    # Generamos una muestra de 10000 elementos con media 0 y desviación estándar 1
    normal_samples = np.random.normal(mean, std, size=k)

    # Graficamos nuestros datos
    fig, ax = plt.subplots()

    plt.hist(normal_samples, bins=50, density=True, color='pink')

    plt.title(f'Distribución para {k} monedas', size=14)
    plt.xlabel('CARA|CRUZ')
    plt.ylabel('Densidad')
    plt.xlim(-3.5,3.5)


    plt.axvline(0, color='#129fef', linestyle='--')
    plt.axvline(np.std(normal_samples), color='#3793ef', linestyle='--')
    plt.axvline(-1*np.std(normal_samples), color='#3793ef', linestyle='--')


normal_dist(50,0,1)
normal_dist(500,0,1)
normal_dist(5000,0,1)
````

#### 2) Muestreo sistematico:

Recuerda que el muestreo sistematico nos sirve para hacer una extraccion de la poblacion con un intervalo constante de como se realizara esta extraccion.

En el ejemplo definiremos una funcion donde tomaremos una muestra del dataset por un intervalo de 3 pasos desde el comienzo hasta el final del mismo

````python
def systematic_sampling(econdata,step):
    # definimos un indece que tome el largo del dataset y salte el numero de pasos que recibe la funcion
    indexes = np.arange(0,len(econdata), step=step)

    # guardamos en systematic_sampling los datos donde esten localizados los indices
    systematic_sampling = econdata.iloc[indexes]

    return systematic_sampling

systematic_sample = systematic_sampling(econdata,30)
systematic_sample
````

#### 3) Muestreo estratificado:

Recordemos que el muestreo estratificado consiste en crear unidades homogeneas exclusivas y a partir de esto crear una muestra aleatoria

````python
# Creamos una variable nueva en el dataset
econdata['estratificado'] = econdata['delegacion'] + ',' + econdata['tipo']
(econdata['estratificado'].value_counts()/len(econdata)).sort_values(ascending=False)
````

Nuestros datos dicen que la proporcion es la siguiente:

1) Hoteles en Cuauhtemoc = 0.5
2) Museos en Cuauhtemoc = 0.2
3) Hoteles en Venustiana Carranza 0.1
4) Mercados en Cuauhtemoc 0.1
5) Mercados en Venustiana Carranza 0.1

````python
def data_estratificada(econdata, nombres_columnas_estrat, valores_estrat, prop_estrat, random_state=None):
    
    df_estrat = pd.DataFrame(columns=econdata.columns) # Creamos un data frame vacío con los nombres de las columnas de econdata
    
    pos = -1
    for i in range(len(valores_estrat)): # iteración sobre los valores estratificados que entran a la funcion 
        pos += 1
        if pos == len(valores_estrat) -1:
            ratio_len = len(econdata) - len(df_estrat) # si es la iteración final calcula el número de valores de salida tenga el mismo número de filas que de entrada
        else:
            ratio_len = int(len(econdata) * prop_estrat[i]) # calcula el número de filas según la proporción deseada
        
        df_filtrado = econdata[econdata[nombres_columnas_estrat] == valores_estrat[i]] # filtra los datos de origen según los valores seleccionados en la estratificación de datos
        df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state) # haz un sample de los datos filtrados usando la ratio que hemos calculado

        df_estrat = pd.concat([df_estrat, df_temp]) # junta las tablas de sample con la estratificada para producir el resultado final

    return df_estrat # Return the stratified, re-sampled data   
````

````python
valores_estrat = ['Cuautémoc,Hotel','Cuautémoc,Museo','Venustiano Carranza,Hotel','Cuauhtémoc,Mercado','Venustiano Carranza,Mercado']
prop_estrat = [0.5, 0.2, 0.1, 0.1, 0.1]
df_estrat = data_estratificada(econdata, 'estratificado',valores_estrat, prop_estrat, random_state = 42)
df_estrat
````

````python
(df_estrat['estratificado'].value_counts()/len(df_estrat)).sort_values(ascending=False)
````

## La media muestral

Antes de empezar con el concepto de media muestral, recordemos los conceptos de **media, moda y mediana**:

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/3/33/Visualisation_mode_median_mean.svg/langes-800px-Visualisation_mode_median_mean.svg.png" width="400">


**Media**: es la suma de los datos divididos entre el numero de datos del conjunto.

$
\displaystyle{\bar{X}=\frac{x_1+x_2+\cdots+x_n}{N}}
$

Por ejemplo si en las clases de matematicas del secundario sacamos **10** en el **primer** trimestre, **7** en el **segundo** y **9** en el **tercero**, la media quedaria tal que:

$
\displaystyle{\bar{X}=\frac{10+7+9}{3}=8.\hat{6}}
$

### Media muestral

La media muestral es lo mismo que la media arimetica, osea el promedio o media de todos los datos.

La media muestral se representa normalmente por el simbolo $\bar X$ y la media poblacional con el simbolo $\mu$, es muy importante que recuerdes que la media muestral y la media poblacional no es lo mismo, que la media muestral surge de una extracion de una poblacion general.

- Media muestral $\bar X$ $\neq$ media poblacional $\mu$

Para dar un ejemplo mas concreto, imaginemos que nos piden estimar la altura promedio de los Colombianos en un rango de edad de los 20 a 30 annos para luego comparar ese promedio muestral con el promedio poblacional del resto de los habitantes.

En este caso extrajimos una muestra, para luego realizar el calculo de un atributo de esta muestra, el cual sera diferente a la media de la problacion general

Para realizar el calculo sumaremos todos los datos y los diviremos entre la cantidad de datos que tenemos.

Podemos imaginar que dentro de una clase tenemos con 20  estudiantes, que serian nuestra muestra, y las edades de los alumnos se reparten entre 28, 24, 25, 23, 38, 52 annos

realizaremos la suma de todas estas edades y las dividiremos por la cantidad de datos que tenemos

$\displaystyle \bar X = \frac{28+24+25+23+38+52}{6}=31.7$

Este resultado seria nuestra media muestral, que representa la edad promedio de los alumnos que elegimos de nuestra clase sobre la poblacion general del curso que serian 20 alumnos  

##  Varianza y desviacion estandar muestral 

La varianza y desviacion estandar nos ayudan a calcular que tan dispersa esta una poblacion o una muestra, que tan separada estarian respecto a la media. 

Mientras mayor sea la varianza, mas distantes estan los datos respecto a la media y en cambio mientras menor sea la varianza, los datos estaran mas pegados respecto a la



Esto lo podemos comprobar si analisamos una poblacion por edades, donde la mayoria de la poblacion tiene 30 annos, entonces tendriamos una varianza pequena, en cambio si tenemos edades de los 10 hasta los 90 annos tendriamos una varianza mucho mayor

### Calculo de la varianza y desviacion estandar

Un punto por el cual te puedes ayudar es que cuando escuches **estandar**, nos estamos refieriendo a una normalizacion, una raiz cuadrada de un componente. Donde la desviacion estandar no es mas ni menos que la raiz cuadrada de la varianza.

Sin embargo no se calcula de la misma manera cuando calculamos la varianza o desviacion estandar de una poblacion o una muestra, ya que tendran unas pequenas diferencias:

$
\begin{vmatrix}
     && varianza                                                                  & & desviacion \;\; estandar\\
\\
Poblacion   && {\displaystyle \sigma^2 = \frac{{\sum _{i=i}^{n}} (x_{i}-\mu)^2}{n}}   & &   {\displaystyle \sigma = \sqrt{\frac{{\sum _{i=i}^{n}} (x_{i}-\mu)^2}{n}}}  & \\
\\
Muestra     && {\displaystyle s^{2} = \frac{\sum _{i=i}^{n} (x_{i}-\mu)^2}{n-1}}          & &   {\displaystyle s = \sqrt{\frac{{\sum _{i=i}^{n}} (x_{i}-\mu)^2}{n-1}}} 
\end{vmatrix}
$

**Diferencias entre caso poblacional y muestral:**

1. Tanto para la varianza y desviacion estandar el para la muestra poblacional es $(\sigma)$ y para la muestral es el simbolo $(s)$  
1. En el calculo muestral, la media o promedio tomara el simbolo $(\bar X)$, pero en el caso poblacional tomara el simbolo $(\mu)$ 
1. El divisor en el caso muestral sera $(N-1)$, pero en el caso poblacional solo sera $(N)$

Hagamos un ejemplo practico usando el mismo ejemplo del caso anterior, donde tenemos una muestra de alumnos, de diferentes edades y ya calculamos su edad promedio

**Media arimetica:**

$\displaystyle \bar X = \frac{28+24+25+38+52}{5}=31.7$

**Varianza muestral:**

${\displaystyle s^{2} = \frac{\sum _{i=i}^{n} (x_{i}-\mu)^2}{n-1}}$

$\displaystyle \bar X = \frac{(28-31.7)^2+(24-31.7)^2+(23-31.7)+(25-31.7)^2+(38-31.7)^2+(52-31.7)^2}{5-1}=31.7$

````python
# Dependencias
import numpy as np
import pandas as pd

def calc_var_std(df,columns):

    index = ['MEAN','VAR','STD','C.V']
    
    # Creamos las listas donde guardaremos los resultados
    mean = []
    var = []
    std = []
    coef = []
    
    # Iteramos los valores de las columnas que queremos calcular: media, varianza, dev est, y coef de variacion
    for i in columns:  
        # Guardamos los valores en las listas creadas
        mean.append(df[i].mean())
        var.append(df[i].var())
        std.append(df[i].std())
        coef.append(df[i].std() / df[i].mean())

    # Damos forma a los datos para el dataframe
    data = np.array([mean,var,std,coef])
    data = data.reshape(len(index),len(columns))

    # Creamos el dataframe
    result = pd.DataFrame(columns=columns, data=data, index=index)
    return result

# Columns son las variables numericas del dataset que queremos calcular
columns = ['sepal-length','sepal-width', 'petal-lenght','petal-width']

calc_var_std(iris,columns) 
````

## Intervalos de confianza

Que son los intervalos de confianza y por que nos sirven tanto para la estadistica, ciencias de datos y en la inteligencia artificial?

El intervalo de confianza describe la variabilidad entre la medida obtenida en una muestra y la medida poblacional (el valor real). Corresponde a un rango de valores, cuya distribución es normal y en el cual se encuentra, con alta probabilidad, el valor real de una determinada variable. Esta «alta probabilidad» se ha establecido por consenso en 95%. Así, un intervalo de confianza de 95% nos indica que dentro del rango dado se encuentra el valor real de un parámetro con 95% de certeza5.

### Nivel de significacion

El nivel de significacion o nivel alfa, es el nivel limite para juzgar si un resultado es o no es estadisticamente significativo. 

Si el valor de significación es menor que el nivel de significación, se considera que el resultado es estadísticamente significativo.

### Interpretacion de un resultado

Veamos ahora como interpretar estos resultados

Intervalo de confianza del 95%:

Al decir que tenemos un nivel de signifiacion al 95%, nos referimos que con un 95% de certidumbre los valores estaran entre dos puntos que abarquen el intervalo.

Un ejemplo puede ser si decimos que la altura de las personas que esquian con un nivel de significacion del 95%, las personas meden entre 160cm y 165cm. De esta forma marcamos el limite inferior y superior del intervalo, quedandonos un 5% de probabilidad de que la pernosa no mida dentro del intervalo, posicionandose por debajo o por encima del intervalo de confianza.

Otro ejemplo que podemos tratar es donde tengamos una lista de alumnos donde podemos comparar con estudiantes que estudiaron 20 horas con otros que estudiaron 5 horas.\
Lo que esperaremos es que la calificacion final sea distinta entre ambas poblaciones, esperando que los que estudiaron 20 horas su califacion oscile entre el 8 y el 10, por otro lado esperamos que los que estudiaron 5 horas oscile entre el 5 y 10 de nota.

Por lo tanto tendriamos un intervalo de confianza mas estrecho para el grupo de las 20 horas: [8,9,10]

Y uno mas largo para los que estudiaron 5 horas: [5,6,7,8,9,10]

### Ejemplo 1

Empecemos por un ejercicio donde no conocemos informacion de nuestra poblacion. Donde trataremos con un 95% de certidumbre calcular entre que valor minimo y maximo se encuentra la poblacion, recordando que del lado izquierdo se encuentra el valor minimo y del lado derecho el maximo.

Si tenemos en cuenta que el ejercicio nos pide saber entre que dos valores esta el $95%$ de la poblacion, sabemos que se encuentra fuera el $5\%$ de la poblacion, que seria el nivel alfa $(\alpha)$, donde la mitad de $\alpha$ abra quedado al lado izquierdo de la poblacion y la otra mitad al lado derecho de la poblacion.

$\alpha = 5\%$

$\displaystyle\frac{\alpha}{2} = 2.5\%$

Al no poseer informacion de la poblacion de la cual estamos calculando, nos guiaremos por una distribucion donde el 0 sera la media de la misma y tendremos una desviacion estandar de 1 (0,1).

Entonces nos planteamos que queremos buscar 2 valores simetricos que cierren el 95% de la poblacion.

Para entonces descubrir el valor de alfa, nos ayudaremos de la tabla de valores Z fijandonos en el valor del porcentaje del intervalo de confianza sumandole el valor de $\displaystyle\frac{\alpha}{2}$


[Tabla de valores Z](https://static.platzi.com/media/public/uploads/z-table_d0a66f42-a480-4942-99f9-438c69248e80.pdf)

$95\% + 2.5\% = 97.5\%$

Entonces vemos que el valor en la tabla Z que tiene da $97.5\%$ es:

$+Z = 1.96$
$-Z = -1.96$

De esta forma sabemos que el intervalo de confianza que nos pide el ejercicio se encuentra entre:

$CI_{95\%}=(-1.96,1.96)$

### Ejemplo 2

En este ejemplo queremos saber la duracion de un cepillo de dientes.

Sabemos que la duracion en dias de un cepillo de dientes se ajusta a la distribucion normal de (28,4) ?Cual es el intervalo de confianza al 80%?

$\mu=28$

$\sigma=4$

En el ejercicio anterior vimos como se resolvia con una media en 0, ahora tenemos que convertir la media a 28 para resolver este ejercicio:

Sabemos que el intervalo de confianza es del 80% y el alfa quedaria en 20%:

$\alpha = 20\%$

Para encontrar el valor de $Z$ tenemos que buscar en nuestra tabla el 80% + 10% $(\alpha /2)$, donde vemos que el valor mas cercano al 90% esta entre 1.2 y 0.008

$Z=1.28$

Ahora procedemos a usar la informacion que tenemos de la duracion de un cepillo sabiendo su media, su desviacion estandar y su valor Z para despejar la formula del valor Z y encontrar los intervalos minimos y maximos de nuestra distribucion

$Z =\displaystyle\frac{x-\mu}{\sigma}$

$Z*\sigma =  x - \mu$

$Z * \sigma + \mu = x$

Despejada la formula para hayar X tenemos en cuenta que tendremos un valor maximo $x_1$ y un valor minimo $x_2$

$x_1 = +Z * \sigma + \mu$

$x_2 = -Z * \sigma + \mu$

Transcribimos los valores a la formula


$x_1 = 1.28 * 4 + 28 = 33.12$

$x_2 = -1.28 * 4 + 28 = 22.88$

## Pruebas de hipotesis

Una prueba de hipotesis o prueba de significacion, nos ayuda a juzgar si existe una diferencia significativa entre el tamano de la muestra y el parametro general, es decir comprobaremos si existe una diferencia entre una teoria y una hipotesis.

Por ejemplo podemos preguntarnos si **en las ciudades donde hace mas frio la gente vive mas,** en el caso de que las personas si vivan mas, estariamos demostrando lo contrario a lo que creemos, ya que deberiamos recurrir a una hipotesis alternativa, dado que lo esperado es que las personas vivan lo mismo independientemente de la temperatura.

Este enunciado seria nuestra hipotesis nula $(h_0)$, que la gente viva lo mismo en un lugar calido como en un lugar frio, porque es nuestra teoria mas normalizada que representaria el centro de nuestra distribucion normal.

En el caso de que las personas vivan mas en lugares calidos como frios, estariamos rechazando nuestra hipotesis nula $(h_0)$, e iriamos a por una hipotesis alternativa que serian nuestra $(h_1)$, estos son nuestros picos tanto del lado izquierdo como del derecho y contemplariamos que si, las personas viven mas dependiendo del clima.

2. Seleccionar el **nivel de significancia**.
3. Seleccionar el **estadistico de prueba.**
4. Formular la regla de decision.
5. Interpretar los resultados y tomar una decision.


#### 1. Establecer uan **hipotesis nula** $(h_0)$ y una **hipotesis alternativa** $(h_1)$

Dado un ejemplo cualquiera, normalmente no diriamos que empezaremos realizando un estudio de ventas, sino que nos enfocariamos en un caso en particular, como puede ser preguntarnos el hecho de que **en verano vendamos mas cerveza por que hace mas calor** (dado el hecho que trabajemos para una empresa que distribuya cerveza).

Pero si efectivamente cuando hace mas calor vendemos mas cerveza, estariamos optando por nuestra hipotesis alternativa, nuestra $h_1$

#### 2. Seleccionar el **nivel de significancia**.

Nos tenemos que plantear a que nivel de significancia queremos realizar nuestro estudio, si queremos verlo al 80%, 90% o al 99%. Definir **con que certidumbre queremos encontrar los parametros que nos interesan y estas diferencias en las distribuciones.**

#### 3. Seleccionar el **estadistico de prueba.**

tenemos: t de student(en poblaciones pequeñas, sin datos de distribución), coeficiente pearson(medir correlación) o ANOVA (comparar la varianza de las distribuciones)

#### 4. Formular la regla de decision.

Si para el 99% de seguridad tenemos una diferencia vamos a hacer un cambio en la empresa, por ejemplo si solo tuvimos un cambio del 68% no queremos hacer ninguna accion.

#### 5. Interpretar los resultados y tomar una decision.

Validado el hecho si existe o no una diferencia, deberemos tomar una decision, lo cual es la razon por la que hacemos un analisis en ciencias de datos.

Por ejemplo si comprobamos que existe una diferencia significativa de ventas en verano a comparacion de las otras estaciones, realizariamos un mes antes de la epoca de calor un incremento en nuestra camapania publicitaria para ganar a los competidores.

El T-value es una relacion de la sennal entre el ruido, las senales seran numeros que nos diran la **diferencia entre nuestras dos muestras** asi que podemos calcular la media de la muestra 1 y la media de lam uestra 2:

$
\text{senal}=\text{diferencia entre la medias de los grupos}=|\bar{x_1}-\bar{x_2}|
$

Y el ruido es la variabilidad que tendremos entre nuestros grupos, asi que lo calcularemos como la raiz cuadrada de la varianza sobre la cantidad de muestras del grupo 1 sumada a la varianza sobre la cantidad de muestras del grupo 2. ($\frac {\sigma}{\sqrt {n}}$ esta formula corresponde al [**error estandar de la media**](https://es.wikipedia.org/wiki/Error_est%C3%A1ndar) o **SEM** por sus siglas en ingles de **S**tandar **E**rror of the **M**ean)

$
\text{ruido}=\text{variabilidad de los grupos}=\sqrt {\displaystyle\frac{(\sigma_{1})^2}{N_{1}}+\displaystyle\frac{(\sigma_{2})^2}{N_{2}}}
$

$
\displaystyle\frac{\text{senal}}{\text{ruido}} = \displaystyle\frac{\text{diferencia entre la medias de los grupos}}{\text{variabilidad de los grupos}} = \displaystyle\frac{|\bar{x_1}-\bar{x_2}|}{\sqrt {\displaystyle\frac{(\sigma_{1})^2}{N_{1}}+\displaystyle\frac{(\sigma_{2})^2}{N_{2}}}}
$

- Un factor es que **mientras mas grande sea la varianza, nuestro T-value sea mas pequeno**.

- Otro factor sera la cantidad de muestras (N) que tomemos, ya que **mientras mas cantidad de muestras, mayor sera nuestra senal** hasta cierto punto dejandonos un T-value mas alto.

### Tipos de pruebas:

#### 1) T de Student

$
Error\;estándar = SE = \displaystyle\frac{\sigma}{\sqrt{N}}
$

Calculamos el error estandar para poder luego hacer el calculo de la varianza

````python
# Variables a estudiar: sepal_length sepal_width

# Calculamos el error estandar
se_length = iris.std()['sepal_length']/np.sqrt(len(iris['sepal_length']))
se_width = iris.std()['sepal_width']/np.sqrt(len(iris['sepal_length']))

se_length,se_width
````

##### Calculamos el ruido de nuestras dos variables

$
SED = \displaystyle \sqrt {\displaystyle\frac{(\sigma_{1})^2}{N_{1}}+\displaystyle\frac{(\sigma_{2})^2}{N_{2}}}
$

````python
sed = np.sqrt((se_length**2) + (se_width**2))
sed
````

##### Calcular el valor de T (T-Value)

**Formula T-Student:**

$t = \displaystyle\frac{(\bar{x_1}-\bar{x_2})}{\displaystyle\sqrt{\frac{{(S_1)}^{2}}{n_1}+\frac{{(S_2)}^{2}}{n_2}}} = \displaystyle\frac{\text{diferencia entre la medias de los grupos}}{\text{variabilidad de los grupos}} = \displaystyle\frac{\text{senal}}{\text{ruido}}$

Abreviacion:

$t = \displaystyle\frac{(\displaystyle\bar{x_1}-\bar{x_2})}{SED}$

````python
# T-Student
t_stat = (iris.mean()['sepal_length'] - iris.mean()['sepal_width']) / sed
t_stat
````

Esto nos dejo un T-value de 36.46

##### Calcular el valor de P

Pasaremos a calcular nuestro P-value que usaremo como valor critico para comprobar si la hipotesis nula de que ambas distribuciones son parecidas se rechaza o se acepta

##### Cuando rechazar una hipotesis


- $\text{Un valor de p > 0.05 significa que la hipotesis nula no se rechaza}$
- $\text{Un valor de p < 0.05 significa que la hipotesis nula se rechaza}$
- $\text{Un valor de p = 0.05 significa que la hipotesis nula se rechaza}$


````python
p = ttest_ind(iris['sepal_length'], iris['sepal_width'])
p
````

````python
if abs(p.statistic) < abs(p.pvalue):
    print('no se rechaza, el T Value es menor que el P Value')
else:
    print('si se rechaza, el T Value es mayor que el P Value')
````

Nuestra teoria inicial era que la distribucion de ambas variables del sepalo eran muy similares, pero como vemos el P-Value que es drasticamente menor que el T-Value, rechazamos la hipotesis nula y optamos por la hipotes alternativa que decia que las distribuciones del largo y ancho el sepalo eran distintas.\
En el caso contrario de que el T Value haya sido menor que el P Value, hubieramos aceptado la hipotesis nula afirmando que eran similares

## Bootstrapping

### Que es el Bootstrapping

Es un metodo de **remuestro con remplazo** de datos dentro de una muestra eleatoria.

Se usa para hallar una aproximacion a la distribucion de la variable analizada.

#### Que es un remuestro con remplazo?

Dado que partimos de una muestra ya extraida de la poblacion, volveremos a tomar una muestra de esta misma muestra.\
Es decir, una muestra de la muestra, donde la muestra original mantendra sus datos originales, pero cada vez que hagamos un remuestreo en este cambien sus datos.

Como vemos en la imagen, imaginemos que los puntos azules representarian a la poblacion de una ciudad con muchos habitantes, donde realizamos una muestra y obtenemos una distribucion no sesgada que nos podria ayudar a sacar una conclusion de la poblacion general.

En el caso de la poblacion de los puntos amarillos imaginemos que es mas pequena, por lo que se aplica la tecnica de Bootstrapping donde hacemos una primer muestra de la poblacion general, donde luego desde esa misma muestra sacamos varias muestras mas.

<img src='https://miro.medium.com/max/1200/1*cmA1PPO4Us6_okb5UMhb0Q.jpeg' width='400'>

```
muestra_original = [1, 2, 3, 4, 5]

# Podemos sacar un mismo valor varias veces
bootstrap_1 = [1, 5, 2, 5, 4]
bootstrap_2 = [5, 2, 1, 4, 4]
```

> 💡 Hay que tener en cuenta que si el muestreo que tenemos tiene 5 datos, el remuestreo tambien debera tener 5 datos al azar de la muestra, si la muestra tiene 10 datos el remuestreo tambien tendra 10 datos de la muestra.

### ¿Por qué usar bootstrapping?

Imagina que estamos estudiando un caso donde no llegamos a una clara conclusion de nuestros experimentos, por ejemplo testear si una droga para curar una enfermedad hace sentir mejor o peor a las personas que la tomen, **experimentos de este tipo tienden a ser costosos y tomar mucho tiempo** y mas aun cuando no podemos controlar parametros del azar que afecten a nuestros experimentos.

Entonces aca es cuando nos preguntamos si hay alguna forma de llegar a una conclusion sin tener que ir a por la opcion **costosa** y que **toma mas tiempo**, que para nuestro ejemplo seria repetir el experimento de las drogas con distintos pacientes un monton de veces.

Por eso en ves de replicar el experimento, aplicamos el Bootstrapping para poder tener una mejor nocion de si los resultados son mas probables o raros de obtener, ademas de ser mas accesible su aplicacion por motivos de computo abaratando los experimentos considerablemente.

> Como recomendación, para hacer bootstrapping, asegúrate de tener al menos 25 registros y correr como mínimo 1000 simulaciones. 🚀

## Validacion cruzada

### Que es la validacion cruzada

La [validación cruzada](https://es.wikipedia.org/wiki/Validaci%C3%B3n_cruzada), o cross validation, se define como una técnica muy utilizada para evaluar los resultados de un análisis estadístico y garantizar que son independientes de la partición entre los datos de entrenamiento y prueba.

Es muy útil para validar que el modelo de aprendizaje automático elegido es el correcto y ayuda a evitar el overfitting.

### Procedimiento
1) Division de los datos de forma aleatoria en k grupos de un tamano similar

2) Se usan k-1 grupos para entrenar el modelo y un ode ellos se usa para validarlo

3) El proceso se rpeite k veces usando un grupo distinto como validacion en cada iteracion

A medida que el proceso se repite, iremos comparando validaciones. Luego, repetiremos k veces el procedimiento hasta llegar al resultado ideal donde nuestra población de entrenamiento y prueba se ajuste el maximo posible.

¿Como medimos el verdadero score de nuestro modelo?

Ayudandonos del [Coeficiente kappa de Cohen](https://es.wikipedia.org/wiki/Coeficiente_kappa_de_Cohen#:~:text=El%20Coeficiente%20kappa%20de%20Cohen,elementos%20cualitativos%20(variables%20categ%C3%B3ricas).) (o Cohen's kappa

Se observa que el score da un aproximado de 78%, muy diferente al 90% que se determina en el caso de que no balanceamos los datos.

````python
from sklearn.model_selection import train_test_split
from sklearn.metrics import cohen_kappa_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

x_train,x_test,y_train,y_test = train_test_split(x_bal, y_bal, test_size=.3, random_state=4)

# Declaramos los modelos decision tree y random forest
modelo_decision_tree = DecisionTreeClassifier()
modelo_random_forest = RandomForestClassifier() 


# Entrenamos el modelo Random Forest con las clases balanceadas
modelo_random_forest.fit(x_train,y_train)
prediccion_random_forest = modelo_random_forest.predict(x_test)
cohen_rf = cohen_kappa_score(prediccion_random_forest, y_test)

# Entrenamos el modelo Decision Tree con las clases balanceadas
modelo_decision_tree.fit(x_train,y_train)
prediccion_decision_tree = modelo_decision_tree.predict(x_test)
cohen_dt = cohen_kappa_score(prediccion_decision_tree, y_test)

# Entrenamos el modelo Decision Tree con las clases no balanceadas
score_dt = modelo_decision_tree.score(x_test,y_test)


print(f'''
Score Random Forest balanceado:   {cohen_rf}
Score Decision Tree balanceado:   {cohen_dt}
Score Decision Tree inbalanceado: {score_dt}'''
)
````

#### Validacion cruzada

El método [K-Fold](https://www.cienciadedatos.net/documentos/30_cross-validation_oneleaveout_bootstrap#k-fold-cross-validation) para (**validacion cruzada**) es también un proceso iterativo. Consiste en dividir los datos de forma aleatoria en k grupos de aproximadamente el mismo tamaño, k-1 grupos se emplean para entrenar el modelo y uno de los grupos se emplea como validación. Este proceso se repite k veces utilizando un grupo distinto como validación en cada iteración. El proceso genera k estimaciones del error cuyo promedio se emplea como estimación final.

````python
from sklearn.model_selection import KFold

# Metodo k-fold
k_fold = KFold(10)

modelo_decision_tree = DecisionTreeClassifier()
modelo_random_forest = RandomForestClassifier() 
````

````python
from sklearn.model_selection import cross_val_score

# Validacion cruzada de ambos modelos balanceados
resultRandomForest = cross_val_score(modelo_random_forest,x_bal,y_bal, cv=k_fold)
resultDecisionTree = cross_val_score(modelo_decision_tree,x_bal,y_bal, cv=k_fold)
# Validacion cruzada para decision tree no balanceado
resultDecisionTree_2 = cross_val_score(modelo_decision_tree,x,y, cv=k_fold)
````

````python
print(f'''
Validacion Cruzada con metodo Random Forest balanceado:
{resultRandomForest}

promedio: {resultRandomForest.mean()}
===================================================
Validacion Cruzada con metodo Decision Tree balanceado:
{resultDecisionTree}

promedio: {resultDecisionTree.mean()}
===================================================
Validacion Cruzada con metodo Decision tre no balanceado:
{resultDecisionTree_2}

promedio: {resultDecisionTree_2.mean()}''')
````

Con esto concluimos el metodo de validacion cruzada para comprobar nuestros modelos, de esta forma podemos ver el acurracy o precision para identificar si el modelo funciona y se ajusta a los resultados predictivos.
